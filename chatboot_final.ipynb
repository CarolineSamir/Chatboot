```python
import json #for reading and writing JSON-encoded data.
# import re
from sklearn.preprocessing import LabelEncoder #for encoding categorical labels with numerical values. 

import numpy as np #for large, multi-dimensional arrays and matrices

from tensorflow.keras.utils import to_categorical # to convert integer-encoded categorical labels into one-hot encoded format.

from tensorflow.keras.preprocessing.text import Tokenizer #for tokenizing text data
# It converts text into sequences of integers, where each integer represents a unique token.

from tensorflow.keras.preprocessing.sequence import pad_sequences # to pad sequences of varying lengths to ensure uniformity.

```

# Reading data

```python
#opens the JSON file in read mode and loads its contents into dictionary named data
with open('intents.json') as file:
  data = json.load(file)
```

```python
data
```

```python
training_examples = [] #A list containing lowercase versions of all training examples extracted from the JSON data.
training_tags = [] #A list containing the tags associated with each training example.
responses = [] #A list containing the responses associated with each training example.

for example in data['intents']:
  for pattern in example['patterns']:
    training_examples.append(pattern.lower())
    training_tags.append(example['tag'])
    responses.append(example['responses'])
```

```python
print(np.shape(training_examples))
print(training_examples)
```

```python
np.shape(training_tags)
```

```python
responses
```

```python
# encode the categorical training tags into numerical labels.
le = LabelEncoder()
le.fit(training_tags)
training_tags = le.fit_transform(training_tags)
print(training_tags)
```

```python
#returns a numpy array representing the one-hot encoded version of the input data (represented as a binary vector).
training_tags = to_categorical(training_tags)
print(training_tags)
```

```python
training_tags
```

```python
# get only unique values from the input training_examples, with duplicates removed.
unique_values = set(training_examples)
```

```python
#tokenize text data and convert it into sequences of integers to be used as input to neural network models.
toknizer =  Tokenizer(num_words= 395)
toknizer.fit_on_texts(training_examples)
sequance = toknizer.texts_to_sequences(training_examples)
```

```python
sequance
```

```python
# calculates the maximum length of sequences in the list sequance.
max(map(len, sequance))
```

```python
#to pad sequences to a uniform length.
padded_sequence= pad_sequences(sequance, truncating='post', maxlen=9)
```

```python
padded_sequence
```

```python
print(np.array(padded_sequence).shape)
print(training_tags.shape)
max_input_length = max(len(seq) for seq in padded_sequence)
print("Maximum input sequence length:", max_input_length)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, GlobalAveragePooling1D
```

```python
chat_model = Sequential() #create a sequential model.

#Embedding particularly for handling text data  
#It is responsible for converting integer-encoded representations of words into dense vectors of fixed size
chat_model.add(Embedding(395, 10, input_length=9)) 
chat_model.add(GlobalAveragePooling1D())
chat_model.add(Dense(300, activation='relu'))
chat_model.add(Dense(150, activation='relu'))
chat_model.add(Dense(200, activation='relu'))
chat_model.add(Dense(38, activation='softmax')) #multi-class classification
```

```python
chat_model.compile(loss = 'categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
chat_model.summary()
#Categorical cross-entropy measures the difference between the true distribution of the target labels and 
#the predicted distribution output by the model.
```

```python
history = chat_model.fit(np.array(padded_sequence), training_tags, epochs=300)
```

```python
chat_model.evaluate(np.array(padded_sequence), training_tags)
```

```python
import pickle

#serializes and saves the toknizer
pickle.dump(toknizer, open("models/toknizer.pkl", 'wb'))
#serializes and saves the le (label encoder) object
pickle.dump(le, open('models/label_encoder.pkl', 'wb'))
#saves the chat model 
chat_model.save('models/chat_model.h5')
```

```python
from tempfile import NamedTemporaryFile
from tensorflow.keras.models import load_model
from flask import Flask, request, render_template

app = Flask(__name__)

model = load_model('models/chat_model.h5')
```

```python
@app.route('/')
def home():
  return render_template('home.html')
```

```python
@app.route('/',methods=['POST'])
def predict():

    #This line retrieves the value associated with the key "user" from the form data of the POST request.
    value=request.form.get("user") 

    #tokenizes the input text, and then pads the sequences to a maximum length of 9
    sent=pad_sequences(toknizer.texts_to_sequences([value]),maxlen=9) 

    # passes the padded sequence through the model to obtain predictions.
    # argmax is used to find the index of the maximum value in the prediction array.
    result=np.argmax(model.predict(np.array(sent)))
    
    #to convert the predicted index back to the original label.
    f_res=le.inverse_transform(np.array(result).reshape(1))

    #a random response from that tag is selected and stored in the output variable.
    output=''
    for tag in data['intents']:
        if tag['tag']== f_res:
            output=np.random.choice(tag['responses'])
            print(f_res)
    #send the result to the user
    return render_template('home.html', prediction_text='{} '.format(output))
```

```python
if __name__ == '__main__':
  app.run()
```
